```python
import numpy as np
import networkx as nx
import random
import copy

from AntColony import *
```

```python
rng_seed = 1234
```

```python
problem_small1 = Problem("./../../data/augerat/A-n32-k05.xml")
problem_small2 = Problem("./../../data/augerat/A-n44-k06.xml")

problem_medium1 = Problem("./../../data/augerat/A-n60-k09.xml")
problem_medium2 = Problem("./../../data/augerat/A-n69-k09.xml")

problem_big1 = Problem("./../../data/uchoa/X-n101-k25.xml")
problem_big2 = Problem("./../../data/uchoa/X-n120-k6.xml")
```

# small1

```python
greedy_s1 = Greedy(print_warnings=True)
greedy_s1.set_problem(10000, 5, problem_small1)
greedy_s1.optimize()
fig1 = plot_solution(greedy_s1, True)
```

```python
antColony_basic_s1 = AntColony(print_warnings=True)
antColony_basic_s1.set_problem(10000, 5, problem_small1)
antColony_basic_s1.optimize(max_iter=10000, print_progress=True, max_time=300, rng_seed=rng_seed)
fig1 = plot_solution(antColony_basic_s1, True)
```

```python
antColony_reduced_s1 = AntColony_Reduced(print_warnings=True)
antColony_reduced_s1.set_problem(10000, 5, problem_small1)
antColony_reduced_s1.optimize(max_iter=10000, print_progress=True, max_time=300, rng_seed=rng_seed)
fig1 = plot_solution(antColony_reduced_s1, True)
```

```python
antColony_divided_s1 = AntColony_Divided(print_warnings=True)
antColony_divided_s1.set_problem(10000, 5, problem_small1)
antColony_divided_s1.optimize(max_iter=10000, print_progress=True, max_time=300, rng_seed=rng_seed)
fig1 = plot_solution(antColony_divided_s1, True)
```

```python
plot_4_solutions(greedy_s1, antColony_basic_s1, antColony_reduced_s1, antColony_divided_s1,
                 labels=True, save_file="small1.png")
```
